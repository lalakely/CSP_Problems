Le problème est de colorer les provinces de Madagascar , avec un contrainte principale : les couleurs des provinces qui se raprochent soient différent

Donc on a comme contrainte :
- Couleur province != province adjacents

```python
import itertools

def solve_csp_gt():
    # provinces (Variables)
    provinces = [
        "Antananarivo", "Antsiranana", "Fianarantsoa", 
        "Mahajanga", "Toamasina", "Toliara"
    ]
    
    # couleurs (Domaine)
    couleurs = ["Rouge", "Vert", "Bleu"]
    
    # adjacences (Contraintes)
    adjacences = {
        "Antsiranana": ["Mahajanga", "Toamasina"],
        "Mahajanga": ["Antsiranana", "Antananarivo", "Toamasina", "Toliara"],
        "Toamasina": ["Antsiranana", "Mahajanga", "Antananarivo", "Fianarantsoa"],
        "Antananarivo": ["Mahajanga", "Toamasina", "Fianarantsoa", "Toliara"],
        "Fianarantsoa": ["Antananarivo", "Toamasina", "Toliara"],
        "Toliara": ["Mahajanga", "Antananarivo", "Fianarantsoa"]
    }

    # On génère toutes les combinaisons possibles (Domaine ^ Variables)
    combinaisons = itertools.product(couleurs, repeat=len(provinces))
    
    tentatives = 0
    for combo in combinaisons:
        tentatives += 1
        # Création d'un dictionnaire d'affectation temporaire
        affectation = dict(zip(provinces, combo))
        
        # Test des contraintes
        est_valide = True
        for province, voisins in adjacences.items():
            for voisin in voisins:
                if affectation[province] == affectation[voisin]:
                    est_valide = False
                    break
            if not est_valide:
                break
        
        # Si le test passe, on a trouvé notre solution
        if est_valide:
            print(f"Solution trouvée après {tentatives} tentatives !")
            return affectation

    return None

solution = solve_csp_gt()

if solution:
    for prov, coul in solution.items():
        print(f"{prov} : {coul}")
else:
    print("Aucune solution trouvée.")
```
